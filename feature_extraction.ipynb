```python
input_file = "../merged_datasets2.csv"
output_file = 'saved_dictionary_6.pkl'
data_cleaner = DatasetParser()

# default output is ta test_cleaned_table_output.txt
```

```python
# process the datasset
# -- parse individual features in the 'features' column
# -- combine duplicate features and standardize names like finish type -> finish types

CONTINUE_AFTER_INTERRUPTION = True

if CONTINUE_AFTER_INTERRUPTION:
    data_cleaner.load_from_file('saved_dictionary_5.pkl')
    
    #ITERATION_START = 112     # if picking up without saved variables
    ITERATION_START = iter # if picking up from the same kernel
else:
    data_cleaner.table = [] # start completely fresh; reset all previously-loaded data
    ITERATION_START = 0

MAX_ITERATIONS = 500
iter = 0
print(f"Processing chunks starting at chunk {ITERATION_START}")
try:
    for chunk in pd.read_csv(input_file, chunksize=10000):
        if iter < ITERATION_START:
            iter = iter + 1
            continue
        chunk.head()
        data_cleaner.format_df_chunk(chunk)
        #data_cleaner.write_table_to_file() # write to plain text file for debug
        iter = iter + 1
        if iter == MAX_ITERATIONS:
            break
except Exception as e:
    raise e
finally:
    print(f"Chunks processed: {iter}; rows saved: {len(data_cleaner.table)}")
    # save current progress to file
    data_cleaner.save_to_file(output_file)
    print(f"Saved to {output_file}")

# save current progress to file. 46m 16.9s for first run.
#data_cleaner.save_to_file('saved_dictionary.pkl') # initial run
```
